```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn import svm

import warnings
warnings.filterwarnings('ignore')
```

```python
gen = pd.read_csv('../dataset/General/Training.csv')
# Identify columns where all values are 0
zero_columns = gen.columns[(gen == 0).all()]

# Drop the identified columns from the DataFrame
gen = gen.drop(zero_columns, axis=1)

gen = gen.rename(columns={'prognosis' : 'diseases'})
gen
```

```python
gen = gen.fillna(0)
gen.isnull().sum()
```

```python
gen['diseases'].value_counts()
```

```python
gen.describe()
```

```python
print(gen.dtypes)
```

```python
gen.groupby('diseases').mean()
```

```python
X = gen.drop(columns='diseases', axis = 1)
Y = gen['diseases']
```

```python
print(X)
```

```python
print(Y)
```

```python
scalar = StandardScaler()
standardize_data = scalar.fit_transform(X)
print(standardize_data)
```

```python
X = standardize_data
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, stratify=Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
classifier = svm.SVC(kernel='linear')
classifier.fit(X_train, Y_train)
```

```python
X_train_prediction = classifier.predict(X_train)
trained_accuracy = accuracy_score(X_train_prediction, Y_train)
print(trained_accuracy)
```

```python
X_test_prediction = classifier.predict(X_test)
tested_accuracy = accuracy_score(X_test_prediction, Y_test)
print(tested_accuracy)
```
