```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score

import warnings
warnings.filterwarnings('ignore')
```

```python
X = df.drop(columns='Outcome', axis=1)
Y = df['Outcome']
print(X)
```

```python
print(Y)
```

```python
scalar = StandardScaler()
Standardized_data = scalar.fit_transform(X)
print(Standardized_data)
```

```python
X = Standardized_data
```

Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

```python
X_train_prediction = classifier.predict(X_train)
trainind_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print(trainind_data_accuracy)
```

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print(test_data_accuracy)
```

```python
input_data = (68,2,1,2,1,1,2,1,1,1,1,1,1,1)
input_data_as_numpy_array = np.asarray(input_data)

input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

std_data = scalar.transform(input_data_reshaped)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0] == 0):
  print('The person do not have Lung Cancer')
else:
  print('The person have Lung Cancer')
```
