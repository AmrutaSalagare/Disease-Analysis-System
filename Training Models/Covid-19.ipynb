```python
#Separating data & labels
X = df.drop(columns='Outcome', axis=1)
Y = df['Outcome']
print(X)
```

```python
print(Y)
```

Data Standardization 

```python
scalar = StandardScaler()
Standardized_data = scalar.fit_transform(X)
print(Standardized_data)
```

```python
X = Standardized_data
Y = df['Outcome']
```

```python
print(X)
print(Y)
```

Train Test Spilt

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
#training the support vecti machine Classifier
classifier.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print(training_data_accuracy)
```

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print(test_data_accuracy)
```

Prediction System

```python
input_data = (0,0,1,0)

#converting input array into numpy array
input_data_as_numpy_array = np.asarray(input_data)

#Reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

std_data = scalar.transform(input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0] == 1):
  print('The person is affected with Covid-19')
else:
  print('The person is not affected with Covid-19')
```
